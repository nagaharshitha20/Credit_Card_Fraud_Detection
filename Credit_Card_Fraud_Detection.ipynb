```python
import pandas as pd
import numpy as np

file_path = "/content/drive/MyDrive/Colab Notebooks/creditcard.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
print("Shape of the dataset:",df.shape)
```

```python
print("Dataset info")
print(df.info())
```

```python
print("\nClass distribution (0 = legit, 1 = fraud):")
print(df['Class'].value_counts())
```

```python
print("\nClass distribution in percentage:")
print(df['Class'].value_counts(normalize=True) * 100)
```

```python
# Separate features (X) and target (y)

X = df.drop("Class", axis=1)   # all columns except Class
y = df["Class"]                # only the Class column

X.head(), y.head()
```

```python
from sklearn.model_selection import train_test_split

# Stratified split to maintain fraud ratio in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train shape:", X_train.shape, y_train.shape)
print("Test shape:", X_test.shape, y_test.shape)

print("\nFraud ratio in train:")
print(y_train.value_counts(normalize=True) * 100)

print("\nFraud ratio in test:")
print(y_test.value_counts(normalize=True) * 100)
```

```python
from sklearn.preprocessing import StandardScaler

# Create the scaler object
scaler = StandardScaler()

# Fit ONLY on training data
scaler.fit(X_train)

# Transform both train and test datasets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Checking shapes to confirm everything is good
print("Scaled X_train shape:", X_train_scaled.shape)
print("Scaled X_test shape:", X_test_scaled.shape)
```

# -------------------------------
# SMALL BALANCED DATASET
# -------------------------------


```python
# Create small balanced dataset: 492 fraud + 492 non-fraud

# Separate fraud and non-fraud rows
fraud = df[df['Class'] == 1]
non_fraud = df[df['Class'] == 0].sample(n=len(fraud), random_state=42)

# Combine them
balanced_df = pd.concat([fraud, non_fraud], axis=0)

# Shuffle the rows
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Show shapes
print("Balanced dataset shape:", balanced_df.shape)

# Check class distribution
print("\nClass distribution:")
print(balanced_df['Class'].value_counts())

print("\nClass distribution in percentage:")
print(balanced_df['Class'].value_counts(normalize=True) * 100)
```

Here we need not to use stratification as we did for full data set because this data set is already balanced with equal fraud and non fraud but still we can keep stratify=y for safety



```python
from sklearn.model_selection import train_test_split

# Separate X and y for balanced dataset
X_balanced = balanced_df.drop("Class", axis=1)
y_balanced = balanced_df["Class"]

# Split into train and test
X_train_small, X_test_small, y_train_small, y_test_small = train_test_split(
    X_balanced, y_balanced,
    test_size=0.2,
    stratify=y_balanced,      # optional but safe
    random_state=42
)

print("Train shape:", X_train_small.shape, y_train_small.shape)
print("Test shape:", X_test_small.shape, y_test_small.shape)

print("\nClass ratio in TRAIN:")
print(y_train_small.value_counts(normalize=True) * 100)

print("\nClass ratio in TEST:")
print(y_test_small.value_counts(normalize=True) * 100)
```

this is a new dataset with 984 rows, not the full dataset.
Scaling must be done on the dataset you actually use for training.So we must:

Create a new scaler

Fit it only on training data

Transform both training and testing data

```python
from sklearn.preprocessing import StandardScaler

# Create scaler
scaler_small = StandardScaler()

# Fit only on training data
scaler_small.fit(X_train_small)

# Transform training and testing data
X_train_small_scaled = scaler_small.transform(X_train_small)
X_test_small_scaled = scaler_small.transform(X_test_small)

# Check shapes
print("Scaled X_train_small shape:", X_train_small_scaled.shape)
print("Scaled X_test_small shape:", X_test_small_scaled.shape)
```

Model building

1.Logistic Reg as baseline model


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create Logistic Regression model
log_reg = LogisticRegression(max_iter=1000, class_weight='balanced')

# Train (fit) on the small scaled training data
log_reg.fit(X_train_small_scaled, y_train_small)

# Predict on the small scaled test data
y_pred_lr = log_reg.predict(X_test_small_scaled)

# Evaluation metrics
print("Accuracy:", accuracy_score(y_test_small, y_pred_lr))

print("\nClassification Report:")
print(classification_report(y_test_small, y_pred_lr))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_small, y_pred_lr))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create the SVM model
svm_model = SVC(kernel='rbf', class_weight='balanced')

# Train the model
svm_model.fit(X_train_small_scaled, y_train_small)

# Predict on the test set
y_pred_svm = svm_model.predict(X_test_small_scaled)

# Evaluation
print("Accuracy:", accuracy_score(y_test_small, y_pred_svm))

print("\nClassification Report:")
print(classification_report(y_test_small, y_pred_svm))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_small, y_pred_svm))
```

RANDOM FOREST (Best model)


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create NEW Random Forest model (fresh variable)
rf_final = RandomForestClassifier(
    n_estimators=500,
    max_depth=None,
    class_weight='balanced',
    random_state=42
)

# Fit on small scaled data
rf_final.fit(X_train_small_scaled, y_train_small)

# Predict on scaled test data
rf_pred_final = rf_final.predict(X_test_small_scaled)

# Evaluation
print("Accuracy:", accuracy_score(y_test_small, rf_pred_final))

print("\nClassification Report:")
print(classification_report(y_test_small, rf_pred_final))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_small, rf_pred_final))
```
